# Visualizing Earnings Based On College Majors


Project Description
-----

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- etc.

By using visualizations, we are going to answer some questions like:

 - Do students in more popular majors make more money?
      - Using scatter plots
 - How many majors are predominantly male? Predominantly female?
      - Using histograms            
 - Which category of majors have the most students?
      -  Using bar plots



```python
import pandas as pd
import matplotlib
%matplotlib inline
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print("\n")
print(recent_grads.head())
print("\n")
print(recent_grads.tail())
print("\n")
recent_grads.describe()
```

```python
row_data_count = recent_grads.shape[0]
print("Raw Data Count:",row_data_count)
```

```python
recent_grads= recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print("Cleaned Data Count:",cleaned_data_count)
```

So far we have read our csv file and dropped the the row which had a missing values.

Pandas, Scatter Plots
-----



```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample Size vs. Median')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample Size vs. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full Time vs. Median')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Share Women vs. Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
```

Do students in more popular majors make more money?

```python
ax = recent_grads.plot(y='Median', x='Total', kind='scatter')
ax.set_title('Total vs. Median')
```

Do students that majored in subjects that were majority female make more money?

```python
ax = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
ax.set_title('Sharewomen vs. Median')
```

Is there any link between the number of full-time employees and median salary?

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Full Time vs. Median')
```

Pandas, Scatter Matrix Plot
----

A scatter matrix plot combines both scatter plots and histograms into one grid of plots and allows us to explore potential relationships and distributions simultaneously. Now let's generate some scatter mix plots!

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size' , 'Median']], figsize=(5,5))
scatter_matrix(recent_grads[['Sample_size' , 'Median','Unemployment_rate']], figsize=(10,10))
```

Pandas, Bar Plots
-----

- Comparing the percentages of women (ShareWomen) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen',legend=False)
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen',legend =False)
```

- Comparing the unemployment rate (Unemployment_rate) from the first ten rows and last ten rows of the recent_grads dataframe.

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate',legend=False)
recent_grads[-10:].plot.bar(x='Major', y='Unemployment_rate',legend =False)
```
